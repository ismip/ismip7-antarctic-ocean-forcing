```python
import os
import xarray as xr
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
```

```python
# Directories
data_path="/media/NAS2/ISMIP7/share_with_modellers"
```

```python
# Load CMIP file
model = "CESM2-WACCM"
scenario = "ssp585"
variable="tf" #'tf' or 'thetao' or "so"

dpath=os.path.join( data_path,"biascorr_cmip",model, "v3",scenario, "zhou_annual_06_nov", "Oyr","thetao_so_tf")
outpath=os.path.join( data_path,"parameterisations","meltMIP")
filename = variable+"_Oyr_CESM2-WACCM_ssp585_r1i1p1f1_ismip8km_60m_"
```

```python
# Create perturbation and safe file 2079-2099
ds = xr.open_mfdataset(os.path.join(dpath, filename+'20*.nc'), combine='by_coords')
ds_mean = ds.sel(time=slice('2079', '2099')).mean(dim='time')
output_file= os.path.join(outpath, filename+"2079-2099.nc")
## safe to dataset
ds_mean.to_netcdf(output_file)
```
