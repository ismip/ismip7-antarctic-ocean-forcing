```python
import os
import xarray as xr
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
```

```python
# Directories
data_path="/media/NAS2/ISMIP7/share_with_modellers"
```

```python
# Load CMIP file
model = "CESM2-WACCM"
scenario = "ssp585"
variable="thetao" #'tf' # or 'thetao'

dpath=os.path.join( data_path,"biascorr_cmip",model, "v3",scenario, "zhou_annual_06_nov", "Oyr","thetao_so_tf")
outpath=os.path.join( data_path,"parameterisations","meltMIP")
filename = variable+"_Oyr_CESM2-WACCM_ssp585_r1i1p1f1_ismip8km_60m_"
```

```python
# Create perturbation and safe file

sel_yrs = [2100, 2200, 2299]

for i,years in enumerate(["2095-2100","2191-2200","2291-2299"]):
    print(years)
    background_file = xr.load_dataset(os.path.join(dpath, filename+years+'.nc'))
    ds = background_file.sel(time=background_file.time.dt.year == sel_yrs[i]) #.drop_dims('time')

    output_file= os.path.join(outpath, filename+str(sel_yrs[i])+".nc")

    ## safe to dataset
    ds.to_netcdf(output_file)
```
